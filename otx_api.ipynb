```python
import requests
import pandas as pd

# Insert your OTX Key (SOAR app will also use this later)
API_KEY = "e53ec18076f64eedaaba3f9e8044462d63e5cc355153d938acc4f8871b91d98c"

headers = {
    "X-OTX-API-KEY": API_KEY
}

url = "https://otx.alienvault.com/api/v1/pulses/subscribed"
response = requests.get(url, headers=headers)

# Safe JSON parsing
try:
    data = response.json()
except Exception as e:
    print("Error parsing JSON:", e)
    print("Raw response:", response.text)
    raise SystemExit()

# Validate API response
if "results" not in data:
    print("‚ùå OTX API did NOT return 'results'. Full response:")
    print(data)
    raise SystemExit("Fix API key or permissions.")

# Extract IPv4 indicators
mal_ips = []
for pulse in data["results"]:
    for ind in pulse.get("indicators", []):
        if ind.get("type") == "IPv4":
            mal_ips.append(ind["indicator"])

# Remove duplicates
mal_ips = list(set(mal_ips))

# Create dataframe
df_otx = pd.DataFrame({"malicious_ip": mal_ips})

df_otx.tail()
```
